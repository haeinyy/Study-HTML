```python
import requests as req
res = req.get('http://ggoreb.com/quiz/pattern.html')
res.encoding = 'utf-8'
text = res.text
print(text)
```

```python
start = text.find('kAe')
end = text.find('-->',start)
text = text[start:end]
text
```
